```python
# Dependencies
import numpy as np
import datetime as dt
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_measurment = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date)==6)
```

```python
#  3. Convert the June temperatures to a list.
june_measurment= june_measurment.all()
print(june_measurment)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_df = pd.DataFrame(june_measurment, columns=['date', 'June Temps'])
june_df
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_df.describe()
```

```python
june_df.set_index(june_df['date'], inplace=True)
print(june_df.to_string(index=False))
```

```python
june_df.plot.hist(bins=12)
plt.tight_layout()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
december_measurement=session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date)==12)
```

```python
# 7. Convert the December temperatures to a list.
december_measurement= december_measurement.all()
print(december_measurement)
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
december_df = pd.DataFrame(december_measurement, columns=['date', 'dec Temps'])
december_df
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
december_df.describe()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
december_df.set_index(december_df['date'], inplace=True)
print(december_df.to_string(index=False))
```

```python
december_df.plot.hist(bins=12)
plt.tight_layout()
```

```python
june_all = june_df.describe()
dec_all = december_df.describe()
all_df = june_all.merge(dec_all, left_index=True, right_index=True)
all_df
```
